# Q1 What does the primary analysis of several categorical features reveal?

```python
## Chi-Square test bcz most features are categorical in nature
from sklearn.feature_selection import chi2
```

```python
df=data[['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact',
       'month', 'day_of_week', 'poutcome', 'y']]
```

```python
x1 = df.drop('y', axis = 'columns')
y1 = df['y']
```

```python
f_score = chi2(x1,y1)
f_score
```

```python
p_value = pd.Series(f_score[1], index = x1.columns)
p_value = p_value.sort_values(ascending = True)
p_value
```

```python
cols = []

for i in p_value.index:
    if p_value[i] <= 0.05:
        print(i, ' :- Reject Null Hypothesis')
    else:
        print(i, ' :- Accept Null Hypothesis')
        cols.append(i)
```

```python
cols
```

```python
data.drop(cols,axis=1,inplace=True)
```
